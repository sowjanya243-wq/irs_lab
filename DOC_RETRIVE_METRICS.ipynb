```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import ipywidgets as widgets
from IPython.display import display, clear_output

# Dataset (Documents)
documents = [
    "Python is a programming language",
    "Machine learning uses Python",
    "Information Retrieval Systems use indexing",
    "Python supports Artificial Intelligence",
    "Database Management System stores data"
]

# Ground Truth (Relevant Documents) - based on document numbers
relevant_docs = [1, 2, 4]

# Text boxes
query_box = widgets.Text(value='Python', description='Query:', layout=widgets.Layout(width='400px'))
threshold_box = widgets.FloatText(value=0.20, description='Threshold:')
button = widgets.Button(description="Calculate", button_style='success')
output = widgets.Output()

def calculate(b):
    with output:
        clear_output()

        query = query_box.value
        threshold = threshold_box.value

        # TF-IDF Vectorization
        vectorizer = TfidfVectorizer()
        tfidf_matrix = vectorizer.fit_transform(documents + [query])

        query_vector = tfidf_matrix[-1]
        document_vectors = tfidf_matrix[:-1]

        similarity_scores = cosine_similarity(query_vector, document_vectors).flatten()

        print("Similarity Scores:")
        for i, score in enumerate(similarity_scores):
            print(f"Document {i+1}: {score:.4f}")

        # Retrieval based on threshold
        retrieved_docs = [i + 1 for i, score in enumerate(similarity_scores) if score >= threshold]
        print("\nRetrieved Documents:", retrieved_docs)

        # Precision, Recall, F1
        retrieved_set = set(retrieved_docs)
        relevant_set = set(relevant_docs)
        relevant_retrieved = retrieved_set.intersection(relevant_set)

        precision = len(relevant_retrieved) / len(retrieved_set) if retrieved_set else 0
        recall = len(relevant_retrieved) / len(relevant_set) if relevant_set else 0
        f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

        print("\nEvaluation Metrics")
        print("------------------")
        print("Precision:", round(precision * 100, 2), "%")
        print("Recall:", round(recall * 100, 2), "%")
        print("F1 Score:", round(f1 * 100, 2), "%")

button.on_click(calculate)
display(query_box, threshold_box, button, output)
```
